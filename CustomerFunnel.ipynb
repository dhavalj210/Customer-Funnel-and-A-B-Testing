Prescriptive and Predictive Analysis

```python
level5_data = df_funnel[df_funnel['KPI'] == 'Lv5_Visitors']
level1_data = df_funnel[df_funnel['KPI'] == 'Lv1_Visitors']
```

```python
grouped_level5 = level5_data.groupby(['Segment', 'City', 'Year', 'Month']).agg({'Value': 'sum'}).reset_index()
grouped_level1 = level1_data.groupby(['Segment', 'City', 'Year', 'Month']).agg({'Value': 'sum'}).reset_index()
```

```python
merged_df = pd.merge(grouped_level5, grouped_level1, on=['Segment', 'City', 'Year', 'Month'], suffixes=('_Level5', '_Level1'))
merged_df['ConversionRate'] = merged_df['Value_Level5'] / merged_df['Value_Level1']
merged_df['Value_Level5'].fillna(0, inplace=True)
merged_df['Value_Level1'].fillna(0, inplace=True)
merged_df['ConversionRate'] = np.where(merged_df['Value_Level1'] != 0,
                                       merged_df['Value_Level5'] / merged_df['Value_Level1'],
                                       np.nan)
```

```python
merged_df
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Segment', y='ConversionRate',hue = 'City',  data=merged_df.dropna())  # Drop rows with NaN values
plt.title('Spread of Conversion Rates Across Various Customer Groups')
plt.xlabel('Segment')
plt.ylabel('Conversion Rate')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='ConversionRate', hue='City', data=merged_df.dropna(subset=['ConversionRate']))
plt.title('Year-on-Year Growth of Conversion Rates by City')
plt.xlabel('Year')
plt.ylabel('Conversion Rate')
plt.legend(title='City', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

As we can see, Kochi has the highest conversion rates of all the cities. This means that we should focus on Kochi more as there are more customers and potential for growth. Then at second comes NCR, we can see a quite big gap between Kochi and NCR conversion rate, proving that we have a wider customer base in Kochi. By this, we can also see that the unkown cities have the least conversion rate, so it would be fine to focus on it the least.

```python
tsa_df = pd.DataFrame(df_funnel.groupby(['Date', 'KPI'])['Value'].sum())
tsa_df = tsa_df.reset_index()
```

```python
tsa_df.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(tsa_df[tsa_df['KPI'] == 'Lv5_Visitors']['Value'])
plt.title('Time Series of Level 5 Customers')
plt.xlabel('Date')
plt.ylabel('Number of Customers')
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.figure(figsize=(12, 6))
plot_acf(tsa_df[tsa_df['KPI'] == 'Lv5_Visitors']['Value'], lags=30, title='Autocorrelation Function (ACF)')
plt.show()

plt.figure(figsize=(12, 6))
plot_pacf(tsa_df[tsa_df['KPI'] == 'Lv5_Visitors']['Value'], lags=15, title='Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
rolling_avg = tsa_df[tsa_df['KPI'] == 'Lv5_Visitors']['Value'].rolling(window=3).mean()

plt.figure(figsize=(12, 6))
plt.plot(tsa_df[tsa_df['KPI'] == 'Lv5_Visitors']['Value'], label='Actual')
plt.plot(rolling_avg, label='Rolling Average (3 months)')
plt.title('Actual vs. Rolling Average Time Series of Level 5 Customers')
plt.xlabel('Date')
plt.ylabel('Number of Customers')
plt.legend()
plt.show()
```

```python
level5_data = tsa_df[tsa_df['KPI'] == 'Lv5_Visitors']
```

```python
model = ARIMA(level5_data['Value'], order=(2, 1, 2))  # Adjust order as needed
results = model.fit()

# Forecast next 3 months
forecast_steps = 3
forecast = results.get_forecast(steps=forecast_steps)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(level5_data['Value'], label='Actual Values')
plt.plot(rolling_avg, label='Rolling Average (3 months)')
plt.plot(pd.date_range(level5_data.index[-1], periods=forecast_steps + 1, freq='M'), [level5_data['Value'].iloc[-1]] + forecast.predicted_mean.tolist(), label='ARIMA Forecast (Next 3 months)')
plt.title('Actual Values, Rolling Average, and ARIMA Forecast of Level 5 Customers')
plt.xlabel('Date')
plt.ylabel('Level 5 Customers')
plt.legend()
plt.show()
```

By the above prediction we can see a decreasing value of the level 5 funnel, indicating that we might loose sum of our existing customers. We need to adopt new ways to prevent this loss of customers.

```python
from statsmodels.stats.power import TTestIndPower

effect_size = 0.03
alpha = 0.05
power = 0.8
current_sample_size = len(abtest)
analysis = TTestIndPower()
result = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, nobs1=current_sample_size, alternative='larger', **{'ratio': None})

if result <= current_sample_size:
    print("Sample size is sufficient.")
else:
    print("Sample size is not sufficient. Additional samples may be needed.")
```

```python
from scipy.stats import ttest_ind

def hypothesis_test(control_group, experiment_group, alpha=0.05):
    t_stat, p_value = ttest_ind(control_group, experiment_group)

    if p_value < alpha:
        return "Experiment Group is Better"
    else:
        return "Control Group is Better"
```
